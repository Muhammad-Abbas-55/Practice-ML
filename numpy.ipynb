```python
import numpy as np

a1 = np.arange(9).reshape(3, 3)
a2 = np.arange(3).reshape(1, 3)

print(a1 , a2)    # print the two arrays

# Broadcasting: means that when performing an operation on two arrays, if the shapes of the arrays are not the same, numpy will automatically broadcast the smaller array to match the shape of the larger array. This allows us to perform operations on arrays of different shapes without having to manually reshape them.
print(a1 + a2)    # add the two arrays element-wise

```
