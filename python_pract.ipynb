```python
# Lets Practice

class Account:

    def __init__(self, balance, account):
        self.acc = account
        self.bal = balance

    def debit(self, debited):
        self.bal -= debited
        print("Debited amount is:", debited)
        print("Remaining balance:", self.bal,"\n")

    def credit(self, credited):
        self.bal += credited
        print("Credited amount is:", credited)
        print("Remaining balance:", self.bal,"\n")


acc1 = Account(5000, 1234567890)
print("Account No : " ,acc1.acc)
print("Balance : ", acc1.bal,"\n")

acc1.debit(50)
acc1.credit(60)
acc1.debit(1000)
acc1.credit(40)


acc2 = Account(1000, 555555555555)
print("Account No : " ,acc2.acc)
print("Balance : ", acc2.bal,"\n")

acc2.debit(100)
acc2.credit(200)
```

```python
# Non-OOP version (simple)
# use variables + functions instead of a class.

def create_account(balance, account_no):
    return {"balance": balance, "account": account_no}


def debit(acc, amount):
    print("Account no is:", acc["account"])
    print("Initial Balance is:", acc["balance"])

    acc["balance"] -= amount
    print("Debited amount is:", amount)
    print("Remaining balance:", acc["balance"] , "\n")


def credit(acc, amount):
    print("Account no is:", acc["account"])
    print("Initial Balance is:", acc["balance"])

    acc["balance"] += amount
    print("Credited amount is:", amount)
    print("Remaining balance:", acc["balance"] , "\n")

# create accounts
acc1 = create_account(5000, 12345678)
acc2 = create_account(1000, 1111112345678)
# print(acc1)

debit(acc1, 50)
credit(acc1, 100)
debit(acc2, 100)
credit(acc2, 500)
```

```python
# private (like) attributes and methods

class Account:

    def __init__(self, acct_no, acct_pass):
        self.account = acct_no
        # to make password to private add __ at the buging of the attr as below
        self.__password = acct_pass

    def reset_pass(self):
        # __password accessible inside the class but not in ouside the class because it is private
        print(self.__password)

    # to make private method add __ at the buging of the method name as below
    def __hello(self):
        print("Hello")

    def welcome(self):
        self.__hello()


a1 = Account("12345","abcd")
print(a1.account)

a1.reset_pass()

# private example
# print(a1.__password)

# private method
# a1.__hello()

# inside welcome method we call the __hello private method 
a1.welcome()
```

```python
# Inheritance
# when one class (child/derived) derives the properties and methods of another class (parent/base) called Inheritance


# class Car:
#     def start_car(self):
#         print("Car Starts")
#     def stop_car(self):
#         print("Car Stops")

# # pass the parent class name in parameter
# class ToyotaCar(Car):
#     def model(self):
#         print("Toyota 123")


# # Direct parent/base class 
# c = Car()
# c.start_car()
# c.stop_car()

# # Accessing child/derived classes 
# t = ToyotaCar()
# t.model()
# t.start_car() 
# t.stop_car() 



# or


# Single level inheritance
class Car:
    color = "black"
    @staticmethod
    def start_car():
        print("Car Starts")
    @staticmethod
    
    def stop_car():
        print("Car Stops")


# pass the parent class name in parameter
class ToyotaCar(Car):
    def __init__(self, name):
        self.name = name
        print(self.name)


# Accessing child/derived classes 
t = ToyotaCar("Toyota 123")

t.start_car()
t.stop_car()

# accessing attribute of parent class
print(t.color)
```

```python
# Types of Inheritance
# Single level Inheritance        base/parent class -> child/derived class        # example in previous notebook
# multi level Inheritance         base/parent class -> child/derived class -> child/derived class
# multiple Inheritance            multiple base/parent class -> inherited by child/derived class


# Multi level color Example
class Car:
    color = "black"
    @staticmethod
    def start_car():
        print("Car Starts")
    @staticmethod
    
    def stop_car():
        print("Car Stops")


# pass the parent class name in parameter
class ToyotaCar(Car):
    def __init__(self, brand):
        self.brand = brand
        print(self.brand)

    def hello(self):
        print("Hello")

class Fortuner(ToyotaCar):
    def __init__(self, type):
        self.type = type
        print(self.type)

# Accessing child/derived classes 
f = Fortuner("Diesel")

f.hello()
f.start_car()
f.stop_car()
```

```python
# Super Method
# super() method is use to access methods of parent class

class Car:
    def __init__(self, type):
        self.type = type

    @staticmethod
    def start_car():
        print("Car Starts")
    
    @staticmethod
    def stop_car():
        print("Car Stops")


# pass the parent class name in parameter
class ToyotaCar(Car):
    def __init__(self, brand, type):
        self.brand = brand
        # print(self.brand)

        super().__init__(type)
        super().start_car()
        super().stop_car()
        
t = ToyotaCar("Toyota 123", "Electric")
print(t.brand)

print(t.type)
```

```python
# Polimorphism Example

# class Complex:
#     def __init__(self, real, img):
#         self.real = real
#         self.img = img

#     def showNumber(self):
#         print(self.real, "i +", self.img, "j")

#     def addNumber(self, num):
#         newReal = self.real + num.real
#         newImg = self.img + num.img

#         return Complex(newReal, newImg)

# c1 = Complex(1, 3)
# c1.showNumber()

# c2 = Complex(4, 9)
# c2.showNumber()

# c3 = c1.addNumber(c2)
# c3.showNumber()




# or



class Complex:
    def __init__(self, real, img):
        self.real = real
        self.img = img

    def showNumber(self):
        print(self.real, "i +", self.img, "j")

    #   __add__   is a Dunder function
    def __add__(self, num):
        newReal = self.real + num.real
        newImg = self.img + num.img

        return Complex(newReal, newImg)
    
    def __sub__(self, num):
        newReal = self.real - num.real
        newImg = self.img - num.img

        return Complex(newReal, newImg)

c1 = Complex(1, 3)
c1.showNumber()

c2 = Complex(4, 9)
c2.showNumber()

# c3 = c1 + c2
c3 = c1 - c2
c3.showNumber()
```

```python
# Dunder function Example

class Order:
    def __init__(self, item, price):
        self.item = item
        self.price = price

    def __gt__(self, odr2):
        return self.price > odr2.price
    
odr1 = Order("chips", 100)
odr2 = Order("tea", 80)

print(odr1 > odr2)      # ture or false
```

```python
# define a circle class to create a circle with radius r using the constructor
# define an area method of the class which calculate  the area of the circle
# define a perimeter() method of the class which allow you to calulate the peremeter of the circle


class Circle:
    def __init__(self, radius):
        self.rad = radius

    def area(self):
        area = 22/7 * (self.rad ** 2)
        return area
    
    def perimeter(self):
        return 2 * 22/7 * self.rad

c1 = Circle(21)
print(c1.area()) 
print(c1.perimeter()) 
```

```python
import os
import chromadb
from sentence_transformers import SentenceTransformer
import google.generativeai as genai

# =========================
# 1. SET GOOGLE API KEY
# =========================
genai.configure(api_key="AIzaSyCnjDFslNNiHOFnPhANp71eG7AMlB-dZLg")

# =========================
# 2. LOAD FILE
# =========================
with open("data.txt", "r", encoding="utf-8") as f:
    text = f.read()

# =========================
# 3. SIMPLE TEXT SPLITTER
# =========================
def split_text(text, chunk_size=100, overlap=20):
    chunks = []
    start = 0

    while start < len(text):
        end = start + chunk_size
        chunks.append(text[start:end])
        start = end - overlap
    return chunks

chunks = split_text(text)

print("Chunks:")
for i, c in enumerate(chunks):
    print(i + 1, c)

# =========================
# 4. EMBEDDINGS MODEL
# =========================
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# =========================
# 5. CHROMA VECTOR DB
# =========================
client = chromadb.PersistentClient(path="./chroma_db")
collection = client.get_or_create_collection(name="docs")

# Store embeddings
for i, chunk in enumerate(chunks):
    embedding = model.encode(chunk).tolist()

    collection.add(
        ids=[str(i)],
        embeddings=[embedding],
        documents=[chunk]
    )

print("\nStored chunks in vector DB")

# =========================
# 6. QUERY FUNCTION
# =========================
def retrieve(query, k=3):
    query_embedding = model.encode(query).tolist()

    results = collection.query(
        query_embeddings=[query_embedding],
        n_results=k
    )

    return results["documents"][0]

# =========================
# 7. GEMINI LLM CALL
# =========================
def ask_llm(context, question):
    prompt = f"""
You are a helpful assistant.

Context:
{context}

Question:
{question}

Answer only from context.
"""

    model = genai.GenerativeModel("gemini-2.5-flash")
    response = model.generate_content(prompt)

    return response.text

# =========================
# 8. RAG PIPELINE
# =========================
def rag(query):
    docs = retrieve(query)
    context = "\n".join(docs)

    answer = ask_llm(context, query)
    return answer

# =========================
# 9. ASK QUESTION
# =========================
query = "Give me only second sentence of the refund policy?"
result = rag(query)

print("\nANSWER:\n", result)
```
